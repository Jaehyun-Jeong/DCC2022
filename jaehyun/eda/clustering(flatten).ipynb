```python
from sklearn.cluster import DBSCAN
from sklearn.decomposition import PCA
import numpy as np

# Use my class
import sys
sys.path.append("../")
from load import Load
```

```python
flatten_loader = Load(flatten=True)
Loader = Load()

flatten_dataset = flatten_loader.numpy_label('../dataset', 'L2_3')
dataset = Loader.numpy_label('../dataset', 'L2_3')
dataset = np.clip(dataset, 0, 1)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4)
clusters = kmeans.fit_predict(flatten_dataset)

zero_idx = [idx for idx, cluster_idx in enumerate(clusters) if cluster_idx == 0]
one_idx = [idx for idx, cluster_idx in enumerate(clusters) if cluster_idx == 1]
two_idx = [idx for idx, cluster_idx in enumerate(clusters) if cluster_idx == 2]
three_idx = [idx for idx, cluster_idx in enumerate(clusters) if cluster_idx == 3]

from matplotlib import pyplot as plt
fig, axes= plt.subplots(40, 10, layout="constrained")

print(np.unique(clusters))
print(np.bincount(clusters))

fig.set_figheight(10)
fig.set_figwidth(10)

for idx, cluster in enumerate(zero_idx[0:100]):
    row = idx % 10
    col = idx // 10

    image = dataset[cluster].transpose(1, 2, 0)
    axes[col][row].imshow(image)
    
for idx, cluster in enumerate(one_idx[0:100]):
    row = idx % 10
    col = (idx // 10) + 10
    
    image = dataset[cluster].transpose(1, 2, 0)
    axes[col][row].imshow(image)

for idx, cluster in enumerate(two_idx[0:100]):
    row = idx % 10
    col = (idx // 10) + 20
    
    image = dataset[cluster].transpose(1, 2, 0)
    axes[col][row].imshow(image)

for idx, cluster in enumerate(three_idx[0:100]):
    row = idx % 10
    col = (idx // 10) + 30
    
    image = dataset[cluster].transpose(1, 2, 0)
    axes[col][row].imshow(image)
```
