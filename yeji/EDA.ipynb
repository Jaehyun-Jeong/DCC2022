```python
from PIL import Image
from os import listdir

path_2 = '/content/drive/MyDrive/Colab Notebooks/DCC/'
dir2_list = listdir(path_2)

image_list = []
img_path = []

for class_name in dir2_list:
  path = '/content/drive/MyDrive/Colab Notebooks/DCC/' + class_name + '/'
  dir_list = listdir(path)

  for filename in dir_list:
    im=Image.open(path+filename)
    image_list.append(im)

image_size_list = [image.size for image in image_list]
print(image_size_list)
```

```python
image_size_dict = {}
for image_size in list(set(image_size_list)):

  image_list = [i for i, size in enumerate(image_size_list) if size[0] == image_size[0] and size[1] == image_size[1]]
  image_len = len(image_list)
  image_size_dict[(image_size)] = image_len
```

```python
x = list(image_size_dict.keys())
count = list(image_size_dict.values())
x1 = []
y1 = []

for i in range(len(x)):
  x1.append(x[i][0])
  y1.append(x[i][1])
```
